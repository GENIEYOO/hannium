```python
import requests
from bs4 import BeautifulSoup
```

```python
def crawl(URL):
    headers={"Accept-Encoding": "false"}
    html=requests.get(URL, headers=headers)
    URLstr=html.content.decode('utf-8')
    return BeautifulSoup(URLstr, "lxml")
```

최신 기사 목록 크롤링

```python
contents=crawl("http://www.hankookilbo.com/lm.aspx").find_all("div","ListCon")
```

```python
for content in contents:
    print(content.get_text().strip())
```

기사 링크 목록 크롤링

```python
links=content.find_all("a")
for link in links:
    print("http://www.hankookilbo.com/"+link.get("href"))
```

링크 별 기사 내용 크롤링

```python
for link in links:
    contents=crawl("http://www.hankookilbo.com/"+link.get('href')).find_all("div","oneSection")
    for content in contents:
        print(content.get_text().strip())
```

```python
bestclick= soup.find(id='bestclick').get_text()
```

```python
bestclick_urls=soup.select("a")
```

```python
for bestclick_url in bestclick_urls[10:]:
    bestclick_url.get("href")
```

```python
html=driver.find_element_by_id('bestclick')
tags=html.find_elements_by_tag_name('a')
print(tags)
```

```python
for tag in tags:
    href=tag.get_attribute('href')
    driver.execute_script("window.open('"+href+"');")
```
